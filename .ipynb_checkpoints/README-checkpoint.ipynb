```python
# %capture installtext
!pip install matplotlib polars numpy pandas
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.axes as axes
import matplotlib.axis as axis
import matplotlib.gridspec as grid

# import polars as plr
import os, re, glob, shutil, gzip

# set the font globally
plt.rcdefaults()
# plt.rcParams.update({'font.family':'sans-serif','font.size':10})
```

```python
file='Book1_new.xlsx'
data=pd.read_excel(file,sheet_name=None)
# Data=plr.read_excel(file,sheet_name=None)
```

```python
def plotDF(df,sheet,col):
    y=df.iloc[:,0]#height
    x=df[col].values
    fig,ax=plt.subplots(1, 1, figsize=(6,4))
    # fig,ax=plt.subplots(1, 1)
    fig.set_figwidth=6
    # fig.set_figheight=4
#     fig.set_facecolor('#BBB')
    
    ax.set_title(sheet)
    ax.grid(True,linewidth=1)
    ax.set_xlabel(col)
    ax.set_ylabel("Height")
    
    
    
    ax.plot(x,y,label=col)
    ax.legend(loc='upper right')
    
    # plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
    # plt.xlim(-3000, 3000)
    # plt.ylim(np.min(y), np.max(y))

    # ax.axes
    # ax.tight_layout()
    plt.savefig(sheet+col)
    plt.show()
    
    # return ax
```

```python


for sheet in list(data.keys()):
    print(sheet)
    df=data[sheet]
    # print(df.head(2))
    for col in list(df.head(0)):
        print(col)
        plotDF(df,sheet,col)
    #     break
    # break
```
